```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot(x, y, std=None,title=None, xlabel=None, ylabel=None, fontsize=20, linestyle='-', color='r'):
    fig = plt.figure()
    plt.xlabel(xlabel, fontsize=fontsize)
    plt.ylabel(ylabel, fontsize=fontsize)
    plt.title(title, fontsize=fontsize+1)
    plt.legend(loc='bottom left')
    
    sns.tsplot(time=x, data=y, color=color, linestyle=linestyle)
    
    plt.xlim([np.min(x)-.5, np.max(x)+.5])
    #plt.ylim([np.min(y)-.5, np.max(y)+.5])
    
    if std is not None:
        plt.errorbar(x=x, y=y, yerr=std, fmt='o')
    
    plt.savefig(title+'.jpg')
    plt.show()
```

```python
demon = [10, 20 , 30, 40]
mean = [3349.6082435, 4391.51489128, 4558.74444345, 4655.8341943]
std = [1334.64405199, 488.158901153, 96.3629654339, 204.179027809]
```

```python
plot(demon, mean, std=std, title='Behavioral Cloning: Demonstrations vs. Reward', xlabel='#of Demonstrations' \
    , ylabel='Mean Reward', color='k', fontsize=15)
```
